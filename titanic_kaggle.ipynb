```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv",delimiter=",",usecols=["Pclass","Sex","Age","SibSp","Parch","Fare"])
df.fillna(0, inplace=True)
df["Sex"] = np.where(df["Sex"] == "female", 0, 1)
X_train=df.to_numpy()

df2=pd.read_csv("train.csv",delimiter=",",usecols=["Survived"])
Y_train=df2.to_numpy()

df3=pd.read_csv("test.csv",delimiter=",",usecols=["Pclass","Sex","Age","SibSp","Parch","Fare"])
df3.fillna(0, inplace=True)
df3["Sex"] = np.where(df3["Sex"] == "female", 0, 1)
X_test=df3.to_numpy()
```

```python
model = Sequential(
    [               
        tf.keras.Input(shape=(6,)),    #specify input size
        ### START CODE HERE ###
        Dense(units=400, activation="relu"),
        Dense(units=192, activation="relu"),
        Dense(units=96, activation="relu"),
        Dense(units=48, activation="relu"),
        Dense(units=24, activation="relu"),
        Dense(units=12, activation="relu"),
        Dense(units=6, activation="relu"),
        Dense(units=1, activation="sigmoid")
        
        
        ### END CODE HERE ### 
    ], name = "my_model" 
)                            
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.001),
)

model.fit(
    X_train,Y_train,
    epochs=200
)
```

```python
Y_test=np.zeros(len(X_test))
for j in range(len(X_test)):
    p=model.predict(X_test[j].reshape(1,6))
    if p>=0.5:
        Y_test[j]=1
    else:
        Y_test[j]=0
Res=np.zeros((418,2))    
for i in range(418):
    Res[i][0]=i+892
    Res[i][1]=Y_test[i]
```

```python
DF=pd.DataFrame(Res)
DF.to_csv("result.csv", index=False,float_format="%d", header=["PassengerId","Survived"])
```
